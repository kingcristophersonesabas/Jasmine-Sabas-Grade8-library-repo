```python
import json

books_data = {
    "books": [
        {
            "id": 2001,
            "title": "Introduction to Python",
            "author": "Juan Dela Cruz",
            "details": {"copies": 5, "available": 3}
        },
        {
            "id": 2002,
            "title": "Data Structures Basics",
            "author": "Maria Santos",
            "details": {"copies": 4, "available": 2}
        },
        {
            "id": 2003,
            "title": "Web Development Essentials",
            "author": "Carlos Reyes",
            "details": {"copies": 6, "available": 6}
        }
    ]
}

print("All Books Data:")
print(json.dumps(books_data, indent=4))
print("\n"+"-"*50)

print("Book Titles:")
for book in books_data["books"]:
    print(f"- {book['title']}")
print("\n"+"-"*50)

print("Book Titles and Availability:")
for book in books_data["books"]:
    print(f"- {book['title']} -> Available: {book['details']['available']}")
print("\n"+"-"*50)

total_books = sum(book["details"]["copies"] for book in books_data["books"])
print(f"Total Number of Books in Library: {total_books}")
```
